```python
# Price prediction using linear regression
df2 = df.set_index(pd.DatetimeIndex(df['DATE'].values))
df2
```

```python
df2 = df2[['CLOSE']]
df2
```

```python
# Create a variable to store the number of days into the future that we want to predict
prediction_days = 1

# Create a new Column called Prediction
df2['Prediction'] = df2[['CLOSE']].shift(-prediction_days)

df2
```

```python
# Create an independent data set (X)
X = np.array(df2.drop(['Prediction'], 1))
# Remove the last n+1 rows of data where n is the prediction_days
X = X[:len(df2) -prediction_days -1]
print(X)
```

```python
# Create a dependent data set (y)
y = np.array(df2['Prediction'])
#Get all of the y values except the last n+1 rows
y = y[:-prediction_days -1]
print(y)
```

```python
# Split the data into 80% training data set and a 20% testing data set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# Use the randomforestRegressor for the model
from sklearn.ensemble import RandomForestRegressor
forest = RandomForestRegressor(n_estimators = 2, random_state = 587)
forest.fit(X_train, y_train)
print(forest.score(X_test, y_test))
```

```python
# Show how close the predicted values and actual values
prediction = forest.predict(X_test)

print(prediction)
print()
print(y_test)
```

```python
# Get the validation data for the model
# Create a variable to store all of the rows in the data set except the last n rows
temp_df2 = df2[:-prediction_days]
# Create a variable to store the independent price value
x_val = temp_df2.tail(1)['CLOSE'][0]
print(x_val)
```

```python
prediction = forest.predict([[x_val]])
# print the price for the next n days
print('The price in', prediction_days, 'day(s) is predicted to be', prediction)
# Print actual value fpr the next n days
print('The actual price was', temp_df2.tail(1)['Prediction'][0])
```
