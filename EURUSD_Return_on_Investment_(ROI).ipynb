```python
df = pd.read_csv('EURUSD_Daily_20060102_20230131.csv')
df
```

```python
df = df.set_index(pd.DatetimeIndex(df['DATE'].values))
df
```

```python
# Ask the user the amount of investment and the date
amount_invested = input('Enter Amount Invested in Dollars: ')
print(amount_invested)
invested_date = input('Enter the date of Investment: ')
print(invested_date)
```

```python
# Fet the low and high price for the day that the user invested to get the price range

# Create variables to store the low and high price column names
col1 = 'LOW'
col2 = 'HIGH'

# Get the low and high price of the asset for the specific day that user invested
price1 = df[col1][invested_date]
price2 = df[col2][invested_date]

# Compute the quantity of the asset that the user would have
quantity1 = int(amount_invested) /  price1
quantity2 = int(amount_invested) /  price2

# Compute the profit range
profit1 = (quantity1 * df[col1][-1]) - int(amount_invested)
profit2 = (quantity1 * df[col2][-1]) - int(amount_invested)

# Print the range amount
print('You would habe made between $', round(profit1, 2), 'and $', round(profit2, 2), 'as of', df['DATE'][-1])
```

```python
# Compute the ROI range
ROT1 = profit1 / int(amount_invested) * 100
ROT2 = profit2 / int(amount_invested) * 100

print('Your ROI would be between', round(ROT1, 2), '% and', round(ROT2, 2), '% as of', df['DATE'][-1])
```
